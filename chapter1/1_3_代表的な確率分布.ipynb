```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats 
```

# Bernoulli distribution

```python
d = stats.bernoulli(0.6)
print(d)
X = d.rvs(100)
X
sum(X)
print(d.pmf(0), d.pmf(1)) # probability mass function of x=0, x=1
plt.bar([0,1],d.pmf([0,1]))
print(d.mean(),d.var())
```

# Categorical distribution

```python
cat_dist = stats.multinomial(1, [0.1, 0.2, 0.3, 0.4]) # sum of parameters is 1

K = len(cat_dist.p)

X_onehot = cat_dist.rvs(100)

X_onehot
X =[np.argmax(x) for x in X_onehot]
X
print(X)
plt.hist(X, bins=range(K+1))


```

```python
X_tmp = np.identity(K)[range(K)]
print(X_tmp)
print(cat_dist.pmf(X_tmp))
plt.bar(range(K), cat_dist.pmf(X_tmp))
```

# Binomial distribution

```python
N = 8
theta = 0.2
bin_dist = stats.binom(N, theta)
X = bin_dist.rvs(100)
X
print('average = ' + str(np.mean(X)))
print('variance = ' + str(np.std(X)**2))
plt.hist(X, range(N+1))
```

```python
ar = np.arange(0, N+1)
plt.bar(ar, bin_dist.pmf(ar))
```

```python
fig, axes = plt.subplots(1,3, figsize=(12,3))
params = [0.1, 0.5, 0.9]
for i, param in enumerate(params):
    bin_dist = stats.binom(N, param)
    axes[i].bar(ar, bin_dist.pmf(ar))
    axes[i].set_title('theta = ' + str(param))
```

# Poisson distribution

```python
poi_dist = stats.poisson(3.0)
X = poi_dist.rvs(100)
np.mean(X)
np.std(X)**2
plt.hist(X, range(30))
```

```python
ar = np.arange(0, 30)
plt.bar(ar, poi_dist.pmf(ar))
```

```python
fig, axes = plt.subplots(1,3,figsize=(12,3))
params = [0.1,1.0,10.0]
for i, param in enumerate(params):
    poi_dist = stats.poisson(param)
    axes[i].bar(np.arange(0,30), poi_dist.pmf(np.arange(0,30)))
    axes[i].set_title('mu = '+str(param))
```

# Gaussian distribution

```python
normal_dist = stats.norm(0.0, 1.0)
X = normal_dist.rvs(1000)
plt.hist(X, bins=10)
normal_dist.ppf(0.1)
```

```python
ls = np.linspace(-3,3,100)
plt.plot(ls, normal_dist.pdf(ls))

```

```python
mu = 0
sigma_list = [0.2, 0.5, 1.0, 2.0]
for sigma in sigma_list:
    normal_dist = stats.norm(mu, sigma)
    plt.plot(ls, normal_dist.pdf(ls),label='mu = ' + str(mu)+', sigma = ' + str(sigma))
plt.legend()
```

# Multivariate Gaussian distribution

```python
mu = [0, 0]
Sigma = [[1.0, 0.0], [0.0, 1.0]]
mvn_dist = stats.multivariate_normal(mu, Sigma)
X = mvn_dist.rvs(1000)
plt.scatter(X[:, 0], X[:, 1])
```

```python
# add weak relation
mu = [0,0]
Sigma = [[1.0, 0.5], [0.5, 1.9]]
mvn_dist = stats.multivariate_normal(mu, Sigma)
X = mvn_dist.rvs(1000)
plt.scatter(X[:,0],X[:,1])
```

```python
x1, x2 = np.mgrid[-3:3:1, -3:3:1]
print(x1)
pos = np.dstack((x1,x2))
plt.contourf(x1,x2,mvn_dist.pdf(pos))
```

# Beta distribution

```python
beta_dist = stats.beta(0.1, 0.1)

X = beta_dist.rvs(1000)

plt.hist(X, bins=10)
```

```python
ls = np.linspace(0,1,100)
plt.plot(ls, beta_dist.pdf(ls))
```

```python
ls = np.linspace(0,1,100)
alpha_list = [0.1,1.0, 2.0]
beta_list = [0.1,1.0,2.0]
for alpha in alpha_list:
    for beta in beta_list:
        beta_dist = stats.beta(alpha,beta)
        plt.plot(ls,beta_dist.pdf(ls),label=str(alpha)+'beta:'+str(beta))
plt.legend()
```

# Dirichlet distribution

```python
dir_dist = stats.dirichlet([0.5,0.5,0.5])
X = dir_dist.rvs(1000)
#plt.figure(figsize=(6,6))
plt.scatter(X[:,0],X[:,1],alpha=0.1)
```

# Gamma distribution

```python
alpha = 1.0
theta = 1.0
gamma_dist = stats.gamma(a=alpha, scale = theta)
X = gamma_dist.rvs(1000)
plt.hist(X,bins=10)
```

```python
mu = 0.3
n1 = stats.norm(1.0, 0.5)
n2 = stats.norm(-1.0,1.0)

N = 10000
Z = np.zeros(N)
X = np.zeros(N)
for n in range(N):
    Z[n] = stats.bernoulli(mu).rvs() # 
    if Z[n] == 0:
        X[n] = n1.rvs()
    else:
        X[n] = n2.rvs()
plt.hist(X, bins=10)
```

```python
def mixture_pdf(mu, n1,n2,x):
    return (1.0-mu)*n1.pdf(x)+mu*n2.pdf(x)

ls = np.linspace(-3,3,100)
plt.plot(ls, mixture_pdf(mu,n1,n2,ls))
```
